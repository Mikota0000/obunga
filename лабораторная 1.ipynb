 promotion_last_5years, на мой взгляд, является категориальной переменной, так как содержит ограниченное количество уникальных значений и состоит из целых чисел.

```
sns.histplot(df['average_montly_hours'], kde=True)
plt.title('Распределение average_montly_hours')
plt.xlabel('average_montly_hours')
plt.ylabel('Частота')
plt.show()
```

```
sns.countplot(x='left', data=df)
plt.title('Распределение left')
plt.xlabel('left')
plt.ylabel('Количество')
plt.show()
```

```
sns.boxplot(x=df['satisfaction_level'])
plt.title('Boxplot satisfaction_level')
plt.xlabel('satisfaction_level')
plt.show()
```

⏰ Распределение average_monthly_hours: Это распределение явно не является нормальным. Мы наблюдаем два пика — около 150 и 250 часов. Для нормального распределения характерен один центральный пик и симметричный спад по обе стороны. В данном случае распределение несимметрично и содержит выбросы.

Распределение left: Это распределение также не может считаться нормальным, так как оно категориальное и отражает количество сотрудников, которые либо остались в компании (0), либо покинули её (1). Нормальное распределение предполагает наличие непрерывных данных, тогда как здесь мы имеем дело с двумя четко определенными категориями.

Boxplot satisfaction_level: По данному графику можно сделать вывод, что данные распределены достаточно равномерно. Боксплот демонстрирует отсутствие выбросов, а само распределение выглядит довольно симметричным, хотя и не обязательно нормальным.

```
df: pd.DataFrame = df.drop('department', axis=1)
df: pd.DataFrame = df.drop('salary', axis=1)

correlation_matrix: pd.DataFrame = df.corr()


sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()
```
